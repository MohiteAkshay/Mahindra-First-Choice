## Calculation % Missing Values

```python
z=0
for i in datasets:
    print('-----------------------'+str(datasets_names[z])+'-----------------------')
    for a in i.columns:   
        
        x=(i[a].isnull().sum())/(i.shape[0])*100
        print( str(a) +' missing % :'+ str(x))
        
    print(' ')
    z=z+1
```

## Deleting columns with missing values >40% missing values

```python
z=0
for i in datasets:
    print('-----------------------'+str(datasets_names[z])+'-----------------------')
    for a in i.columns:
        x=(i[a].isnull().sum())/(i.shape[0])*100
        if x>40:
            i.drop(a,axis=1,inplace=True)
            print(str(a))
    print(' ')
    z=z+1
```
